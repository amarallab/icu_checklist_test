```python
mock_users = [2, 3, 4,5, 7, 8, 20, 51]
```

```python
df_data = df_data[~df_data['user_id'].isin(mock_users)]
df_user = df_user[~df_user.index.isin(mock_users)]
```

```python
adopter_series =  df_data.groupby(df_data['date'].map(lambda x: x.date())).user_id.unique()
```

```python
adopter_df = pd.DataFrame(adopter_series)
adopter_df.set_index(pd.DatetimeIndex(adopter_df.index), inplace=True)
adopter_df = adopter_df.resample('D', fill_method='pad')
```

```python
cum_id = []
adopter_num = []
for i, row in adopter_df.iterrows():
    adopter_list = list(set([item for sublist in adopter_df[:i].values.flatten() for item in sublist]))
    cum_id.append(adopter_list)
    adopter_num.append(len(adopter_list))
```

```python
adopter_df['cum_id'] = pd.Series(cum_id, index=adopter_df.index)
adopter_df['cum_num'] = pd.Series(adopter_num, index=adopter_df.index)
```

```python
fig, ax = plt.subplots(figsize=(18,10))

ax.plot(adopter_df.index, adopter_df.cum_num, lw=4)
# ax.set_ylim([0, len(df_user.index)])
ax.axhline(y=18, color='gray', lw=3)
ax.set_ylim([0, 20])
ax.set_ylabel('Fellows and Nurse Practioners')
ax.set_xlabel('Time')
labels = ax.get_xticklabels()
plt.setp(labels, fontsize=20)
lables=ax.get_yticklabels()
plt.setp(lables, fontsize=20)
ax.text(0.05, 0.91, 'Total number of fellows and nurse practicioners in MICU', color='grey', fontsize=18, transform=ax.transAxes)
save_name = '../result/iphoneapp/adopters_iphone_app_fel_only.pdf'
plt.savefig(save_name, transparent=True, bbox_inches='tight',dpi=300)
```
